```python
import pandas
```

```python
df = pandas.read_excel('s3://isat252-mensah/house_price.xls')
df.plot(x='built_in',y='price')
```

## Line Chart 

avg prices per year 

```python
avg_price_per_year =df.groupby('built_in').mean()['price']
avg_price_per_year.plot()
```

## Histogram
to show distribution of prices

```python
df['price'].hist()
```

## bar chart 
avg prices per house type 

```python
avg_price_per_type = df.groupby('house_type').mean()['price']
avg_price_per_type.plot.bar()
```

# pie graph 
number of records per house type

```python
count_per_house_type = df["house_type"].value_counts()
count_per_house_type.plot.pie()
```

# scatter plot
the relationship between price and area 

```python
df.plot.scatter(x='area',y='price')
```
